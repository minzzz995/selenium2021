```python
import os
import json
from PIL import Image
import datetime

# images 디렉토리 및 파일명
wd = os.getcwd()
print(wd)
os.chdir('image')
wd = os.getcwd()
print(wd)
names = os.listdir(wd) # 리스트로 파일 이름 저장
names.sort() # 오름차순 정렬
fileEx = r".jpg" # .jpg 파일만 선택
names = list(file for file in names if file.endswith(fileEx))
names

# 
l = list()
for name in names:
    d = dict()
    t = datetime.datetime.now()
    t = t.strftime('%Y-%m-%d %H:%M:%S')
    
    d["license"] = 9
    d["file_name"] = name
    d["coco_url"] = "http://192.168.0.219:8007/" + name
    image = Image.open(name)
    d["height"] = image.size[0]
    d["width"] = image.size[1]
    d["data_captured"] = t
    d["flickr_url"] = "http://192.168.0.219:8007/" + name
    d["id"] = int(name[5:-4])
    d["url"] = "http://192.168.0.219:8007/" + name
    d["rights_holder"] = ""
    
    l.append(d)
    


## 원본에서 "images"만 수정 및 저장

# 경로
os.chdir("..")
# 원본 > os.chdir("annotations")

# load
with open("person_keypoints_iwaz2021 copy.json", "r") as f:
# 원본 > with open("person_keypoints_iwaz2021.json", "r") as f:
    j = json.load(f)

# revise    
j["images"] = l
# write
with open("person_keypoints_iwaz2021.json", "w", encoding = "utf-8") as make_file:
    json.dump(d2, make_file, indent="\t")
#print(json.dumps(j["images"], indent="\t"))
```
